#### Plot the distribution of ethnicity - using the same data set in tutorial

```python
# write code here
import pandas as pd
import seaborn as sns
data = pd.read_csv("BWSI_Dataset1.csv") # importing data using pandas 
data.head(20) # view data set
```

#### Plot "hours of TV time" distribution by sex 

```python
# write code here
males = data[data['Sex'] == 1]
females = data[data['Sex'] == 2]

plt.figure(2, figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.countplot(males['HoursTV'])
plt.subplot(1, 2, 2)
sns.countplot(females['HoursTV'])
plt.show()
```

#### Explore the relationship between BMI and age. Does there appear to be an association?

```python
# write code here
plt.figure(1, figsize=(24, 6))
sns.violinplot(x=data['Age'], y=data['BMI'])
```

#### Explore the relationship between BMI and sex by ethnicity. What ethnicity and sex has the highest median BMI? 

```python
# write code here
rc = sns.FacetGrid(data, row='Sex', col='Ethnicity')
rc.map(sns.boxplot(), 'BMI')
```
